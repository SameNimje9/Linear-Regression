Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Importing the dataset

```python
dataset = pd.read_csv("Position_Salaries1.csv")
x=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

```python
y=y.reshape(len(y),1)
print(y)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
sc_y=StandardScaler()
x=sc_x.fit_transform(x)
y=sc_y.fit_transform(y)
```

```python
print(x)
```

```python
print(y)
```

Training SVR model on the whole dataset 

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(x,y)
```

Predicting new results

```python
sc_y.inverse_transform(regressor.predict(sc_x.transform([[6.5]])).reshape(-1,1))
```

Visualing the SVR Results

```python
plt.scatter(sc_x.inverse_transform(x),sc_y.inverse_transform(y),color='red')
plt.plot(sc_x.inverse_transform(x),sc_y.inverse_transform(regressor.predict(x).reshape(-1,1)),color='green')
plt.title('SVR')
plt.xlabel('position level')
plt.ylabel('Salary')
plt.show()
```

```python
x_grid= np.arange(min(sc_x.inverse_transform(x)),max(sc_x.inverse_transform(x)),0.1)
x_grid=x_grid.reshape(len(x_grid),1)
plt.scatter(sc_x.inverse_transform(x),sc_y.inverse_transform(y),color='red')
plt.plot(x_grid,sc_y.inverse_transform(regressor.predict(sc_x.transform(x_grid)).reshape(-1,1)),color='green')
plt.title('SVR')
plt.xlabel('position level')
plt.ylabel('Salary')
plt.show()
```
